```python
#NUMBERS IN PYTHON
#ADDITION

num_1=2
num_2=4
num_1+num_2
print(num_1+num_2)

num_3=2.2
num_4=4.2
sum_3_4=num_3+num_4
print(sum_3_4)

#SUBTRACTION

NUM_1=2
NUM_2=4.7
DIFF_1_2=NUM_1-NUM_2
print(DIFF_1_2)


#MULTIPLICATION

num_5=2
num_6=5.2
mult_5_6=num_5*num_6
print(mult_5_6)


#DIVISION
#NORMAL DIVISION

NUM_3=4
NUM_4=5
div_3_4=NUM_3/NUM_4
print(div_3_4)

#FLOOR DIVISION (It always goes down to the largest integer that is less than or equal to the result.)
div_3_4_quotient=NUM_3//NUM_4
print(div_3_4_quotient)

div_1_2_quotient=num_1//num_2
print(div_1_2_quotient)

num_7=10
num_8=3
div_7_8_quotient=num_7//num_8
print(div_7_8_quotient)


#COMPLEX NUMBERS

comp_a=complex(2,3)
print(comp_a)

comp_b=2+3j
print(comp_b)


#IDENTIFYING THE TYPE OF NUMBER

type(num_3)

#one at one time

type(sum_3_4)

type(div_3_4_quotient)
```
