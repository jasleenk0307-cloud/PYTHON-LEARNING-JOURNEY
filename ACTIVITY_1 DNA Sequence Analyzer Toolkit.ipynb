**FIND REVERSE COMPLEMENT OF DNA SEQUENCE**

```python
DNA='GTTATGCCCTAAGCTGAGCTA'
# DNA IS WRITTEN IN 5' TO 3'
seq=DNA.replace('A','t')
seq=seq.replace('T','a')
seq=seq.replace('C','g')
seq=seq.replace('G','c')
seq=seq.upper()
seq=seq[::-1]  # for reverse the order
print(seq)
```

**CALCULATE GC CONTENT NEED TO KNOW THERMAL STABILITY, REPLICATION EFFICIENCY, AND GENOMIC CHARACTERISTICS**

```python
count_G=DNA.count('G')
count_C=DNA.count('C')
count_GC=count_G+count_C
total_length_DNA=len(DNA)
GC_content=(count_GC/total_length_DNA)*100
GC_content=round(GC_content,2)
print(GC_content)
print(f" GC_content of this DNA fragment is {GC_content}")
```

**DNA Sequence Analyzer Toolkit**

```python
DNA_sequence="GCGCTGAATTCATGACGGTAAAAG"
DNA_seq_length=len(DNA_sequence)
G_content= DNA_sequence.count('G')
C_content= DNA_sequence.count('C')
GC_content=round(((G_content+C_content)/DNA_seq_length)*100)
DNA_replace=DNA_sequence.replace('A','t')
DNA_replace=DNA_replace.replace('T','a')
DNA_replace=DNA_replace.replace('C','g')
DNA_replace=DNA_replace.replace('G','c')
DNA_replace=DNA_replace.upper()
DNA_replace=DNA_replace[::-1]
RNA_sequence=DNA_sequence.replace('T','U')
DNA_count_A=DNA_sequence.count('A')
DNA_count_G=DNA_sequence.count('G')
DNA_count_C=DNA_sequence.count('C')
DNA_count_T=DNA_sequence.count('T')
start_codon=RNA_sequence.find('AUG')
EcoR1_sequence='GAATTC'
EcoR1_sequence=DNA_sequence.find(EcoR1_sequence)
print(f"LENGTH OF DNA SEQUENCE IS {DNA_seq_length}")
print(f"GC CONTENT OF DNA SEQUENCE IS {GC_content}")
print(f"COMPLEMENTARY DNA SEQUENCE IS {DNA_replace}")
print(f"TRANSCRIBED RNA SEQUENCE IS {RNA_sequence}")
print(DNA_count_A)
print(DNA_count_G)
print(DNA_count_C)
print(DNA_count_T)
print(f"START CODON (AUG) FOUND AT POSITION:{start_codon}")
print(f"EcoR1 SEQUENCE IS AT {EcoR1_sequence} POSITION")
```
