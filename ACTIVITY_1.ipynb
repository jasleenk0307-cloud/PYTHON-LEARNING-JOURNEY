**FIND REVERSE COMPLEMENT OF DNA SEQUENCE**

```python
DNA='GTTATGCCCTAAGCTGAGCTA'
# DNA IS WRITTEN IN 5' TO 3'
seq=DNA.replace('A','t')
seq=seq.replace('T','a')
seq=seq.replace('C','g')
seq=seq.replace('G','c')
seq=seq.upper()
seq=seq[::-1]  # for reverse the order
print(seq)
```

**CALCULATE GC CONTENT NEED TO KNOW THERMAL STABILITY, REPLICATION EFFICIENCY, AND GENOMIC CHARACTERISTICS**

```python
count_G=DNA.count('G')
count_C=DNA.count('C')
count_GC=count_G+count_C
total_length_DNA=len(DNA)
GC_content=(count_GC/total_length_DNA)*100
GC_content=round(GC_content,2)
print(GC_content)
print(f" GC_content of this DNA fragment is {GC_content}")
```
