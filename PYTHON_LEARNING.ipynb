```python
# string analysis in python

protein_seq1='VLSPADKTNVK'
protein_seq1_len=len(protein_seq1)      #len() allows us to calculate length of given string
print(protein_seq1_len)



# SPLICING: FORWARD INDEX SPLICING
# string[start : stop : step]
# start = index to begin from
# stop = index before which to stop
# step = how many positions to move

print(protein_seq1[1:5])

print(protein_seq1[1:(5+1)])

print(protein_seq1[1:(5+2)])

print(protein_seq1[2:(5+3)])    # same as protein_seq1[2:8]

print(protein_seq1[1:(5+1):2])   #Start at index 1 and move by 2 positions each time.

print(protein_seq1[2:(7+1):3])

# Omitting the Start Index

print(protein_seq1[:(5+1)])

print(protein_seq1[2:1])   # Since 2 is already beyond 1, nothing can be selected.

print(protein_seq1[:])     # Both start and stop are omitted. This is often used to create a copy of a string


# FLIPPING A SEQUENCE

print(protein_seq1[::1])      #Starts at the beginning, ends at the end, step size = 1.
print(protein_seq1[::-1])     #Step size = -1, so Python reads the string from the end to the beginning.
print(protein_seq1[::-2])
```

# **STRING ANALYSIS IN PYTHON**

CASE CHANGE,
TITLE MODE, SENTENCE MODE


```python
# CASE CHANGE IN PYTHON

DNA_upper='AATTGCGCTTA'
DNA_lower='aactgcgcata'

# to lower
DNA_upper_to_lower=DNA_upper.lower()
print(DNA_upper_to_lower)
print(f"DNA upper is {DNA_upper} and after applying .lower() it became {DNA_upper_to_lower}")

# to upper
DNA_lower_to_upper=DNA_lower.upper()
print(DNA_lower_to_upper)
print(f"DNA upper is {DNA_lower} and after applying .lower() it became {DNA_lower_to_upper}")

# title mode
simulation_name='molecular dynamics simulation'
simulation_name_formatted=simulation_name.title()
print(f"simulation name before: {simulation_name} and after applying .title(), it became {simulation_name_formatted}.")

# sentence mode- capitalize
to_capitalize='learning python is fun through udemy'
to_capitalize_format=to_capitalize.capitalize()
print(f"before format: {to_capitalize}, after applying, capitalize() it became: {to_capitalize_format}")


```

```python
# REPLACE: IN BIOLOGY, WE CAN USE IT TO CREATE MUTANTS OF WILD TYPE DNA/RNA/PROTEIN SEQUENCE
protein_WT='VLSPADKTNVWWWYYY'
mutant_1=protein_WT.replace('W','M')
print(mutant_1)

mutant_2=protein_WT.replace('W','M',1)
print(mutant_2)

mutant_3=protein_WT.replace('W','M',2)
print(mutant_3)

general_example='I am enjoying python programming and i will use it to analyse biological data'
general_example_modified=general_example.replace('analyse','analyze')
print(general_example_modified)
```

```python
# FIND- search for the occurance or position of substring
protein='VLSPADKTNWWWWYYYY'
proline_occurance=protein.find('P')
print(proline_occurance)
tryptophan_occurance=protein.find('W')
print(tryptophan_occurance)

# COUNT- number of times it occured
tryptophan_count=protein.count('W')
print(tryptophan_count)
```

```python
# SPLIT METHOD FOR STRING ANALYSIS- TO DIVIDE A SENTENCE INTO SMALLER PIECES

string_to_split="BIOLOGY IS INTERESTING, I AM USING PYTHON TO ANALYZE BIOLOGICAL DATA."
splitting_string_attempt1=string_to_split.split()
print(splitting_string_attempt1)

splitting_string_attempt2=string_to_split.split(',')
print(splitting_string_attempt2)

splitting_string_attempt3=string_to_split.split('ANALYZE')
print(splitting_string_attempt3)

# JOIN

splitting_string_attempt1_recombine=' '.join(splitting_string_attempt1)
print(splitting_string_attempt1_recombine)

splitting_string_attempt2_recombine=string_to_split.split(':')
print(splitting_string_attempt2_recombine)

# STRIP- METHOD ALLOWS US TO REMOVE CUSTOM PATTERN THAT OCCUR EITHER AT THE BEGINNING OF THE STRING(LEADING) OR AT THE END OF A STRING(TRAILING)

temp_1='python'
temp_2=' python '
print(temp_1)
print(temp_2)

temp_2_cleaned=temp_2.strip()
print(temp_2_cleaned)

temp_2_clean=temp_2.rstrip()   # rstrip means trailing
print(temp_2_clean)

temp_2_cleaned_l=temp_2.lstrip()
print(temp_2_cleaned_l)

# REPLACE
temp_3=" Python is an interesting programming language it is fun to learn it "
print(temp_3.replace(" ",""))
```
